```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

dataset_char = pd.read_csv("A_Z Handwritten Data.csv")
dataset_char.rename(columns={'0':'label'}, inplace=True)
#Splite data the X - Our data , and y - the prdict label
X_char = dataset_char.drop('label',axis = 1).to_numpy()
y_char = dataset_char['label'].to_numpy()
```

```python
X_char.shape
```

```python
from sklearn.decomposition import PCA
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_char =X_char.reshape((372450, 28 * 28))
X_char = X_char.astype('float32') / 255

pca = PCA(n_components=2)
train_images_pca = pca.fit_transform(X_char)

plt.scatter(train_images_pca[:, 0], train_images_pca[:, 1], c=y_char)
plt.colorbar()
plt.show()
```

```python
from sklearn.datasets import fetch_openml
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(n_components=2, random_state=42) # Create a TSNE object with 2 dimensions and a fixed random state
X_tsne = tsne.fit_transform(X_char) # Transform the data to 2 dimensions

# Plot the data with different colors for different digits
plt.figure(figsize=(10, 10)) # Create a figure with a large size
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y_char, cmap='tab10') # Scatter plot the points with colors according to the labels
plt.colorbar() # Add a colorbar to show the mapping of colors and digits
plt.xlabel('t-SNE 1') # Add a label for the x-axis
plt.ylabel('t-SNE 2') # Add a label for the y-axis
plt.title('Character data visualized using t-SNE') # Add a title for the plot
plt.show() # Show the plot+










































































































---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
```
